## **Solution 2**

```python
import os
import pandas as pd

def read_file(file_path, old_files):
  all_files = [f for f in os.listdir(file_path) if f.startswith("Sales_Data") and f.endswith(".csv")]
  new_files = [f for f in all_files if f not in old_files]
  if new_files:
    combined_data = pd.DataFrame()
    for file_name in new_files:
            file_path = os.path.join(file_path, file_name)
            df = pd.read_csv(file_path)
            combined_data = pd.concat([combined_data, df], ignore_index=True)
    old_files.extend(new_files)
    return combined_data, old_files
  else:
      print("No new files to process.")
      return None, old_files

file_path = "/content"
old_files = []
combined_data, updated_old_files = read_file(file_path, old_files)
old_files = updated_old_files
```
